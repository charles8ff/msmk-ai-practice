# Ejercicio 1: Dado el siguiente DataFrame de pandas, identifique las celdas con datos faltantes.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': [1, 2, np.nan, 4],
    'B': [5, np.nan, np.nan, 8],
    'C': [9, 10, 11, 12]
})
print(df.isnull())
```



# Ejercicio 3: Dado el DataFrame anterior, reemplaza los valores faltantes en las columnas 'A' y 'B' con la media de los valores existentes en sus respectivas columnas.

```python
df['A'].fillna(df['A'].mean(), inplace= True)
df['B'].fillna(df['B'].mean(), inplace= True)
df['B'].fillna(df['B'].mean(), inplace= True)
```
